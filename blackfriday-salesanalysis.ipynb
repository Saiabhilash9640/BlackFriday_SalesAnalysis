```python
sales=data.drop(['User_ID', 'Product_ID','Product_Category_3','Product_Category_2'],axis=1)
```

```python
sales
```

```python
sales.isna().sum()
```

```python
sales.Age.value_counts()
```

```python
sales=pd.get_dummies(sales,drop_first=True,dtype='int')
```

```python
sales.columns
```

```python
occu=pd.get_dummies(sales['Occupation'],drop_first=True,prefix='Occ',dtype='int')
occu
```

```python
prod=pd.get_dummies(sales['Product_Category_1'],drop_first=True,prefix='prod',dtype='int')
prod
```

```python
sales.drop(['Occupation','Product_Category_1'],axis=1,inplace=True)
```

```python
sales=pd.concat([sales,occu,prod],axis=1)
```

```python
x=sales.drop(['Purchase'],axis=1)
x
```

```python
y=sales.Purchase
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
z=sc.fit_transform(x['Stay_In_Current_City_Years'].values.reshape(-1,1))
```

```python
z=pd.DataFrame(z,columns=['Stay_In_Current_City_Years'])
```

```python
x.drop(['Stay_In_Current_City_Years'],axis=1,inplace=True)
```

```python
x.insert(0,'Stay_In_Current_City_Years',z)
```

```python
x
```

```python
x.shape
```

```python
x.columns=x.columns.astype('str')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=True)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
lm=LinearRegression()
rfe=RFE(lm,n_features_to_select=15)
rfe.fit(x_train,y_train)
rfe.support_
```

```python
x_train.columns[rfe.support_]
```

```python
x_train_rfe=x_train[x_train.columns[rfe.support_]]
```

```python
x_train_rfe.shape
```

```python
import statsmodels.api as sm
x_train_lm=sm.add_constant(x_train_rfe)
x_train_lm
```

```python
lr=sm.OLS(y_train,x_train_lm).fit()
lr.summary()
```

```python
x_train_lm1=sm.add_constant(x_train)
```

```python
lr=sm.OLS(y_train,x_train_lm1).fit()
lr.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['Features']=x_train.columns
vif['vif']=[variance_inflation_factor(x_train.values,i) for i in range(x_train.shape[1])]
vif
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score,mean_squared_error
model=Ridge()
lm=model.fit(x_train,y_train)
```

```python
y_pred=lm.predict(x_test)
print('mean_squared_error ',mean_squared_error(y_test,y_pred))
print('r_square score ',r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Lasso
model=Lasso()
lm1=model.fit(x_train,y_train)
y_pred=lm1.predict(x_test)
print('mean_squared_error ',mean_squared_error(y_test,y_pred))
print('r_square score ',r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import ElasticNet
model=ElasticNet()
lm2=model.fit(x_train,y_train)
y_pred=lm2.predict(x_test)
print('mean_squared_error ',mean_squared_error(y_test,y_pred))
print('r_square score ',r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor(n_estimators=100,max_depth=5,min_samples_split=2,learning_rate=0.1)
lm3=model.fit(x_train,y_train)
y_pred=lm3.predict(x_test)
print('mean_squared_error ',mean_squared_error(y_test,y_pred))
print('r_square score ',r2_score(y_test,y_pred))
```

```python
import seaborn as sns
sns.distplot(y_test-y_pred,bins=20)
plt.title('Error terms')
```
